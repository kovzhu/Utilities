```python
def generate_raw_url(relative_link):
    """
    Generates a raw GitHub URL from a relative link.

    Args:
        relative_link (str): The relative GitHub link.

    Returns:
        str: The corresponding raw GitHub URL.
    """
    base_url = "https://raw.githubusercontent.com"
    raw_path = relative_link.replace("blob/", "")
    # Remove the leading slash if it exists
    # if raw_path.startswith('/'):
    #     raw_path = raw_path[1:]
    full_url = urllib.parse.urljoin(base_url, raw_path)
    return full_url
```

```python
def get_economist_string(url):
    """
    Extracts the string '01_economist' from a GitHub URL.

    Args:
        url (str): The GitHub URL.

    Returns:
        str: The string '01_economist' if found, or None if not found.
    """
    try:
        parsed_url = urllib.parse.urlparse(url)
        path_segments = parsed_url.path.split('/')

        # Find "01_economist" in the path segments.
        if "01_economist" in path_segments:
            return "01_economist"
        else:
            return None

    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```
